biGRU

```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report

input_dim_1 = 12
input_dim_2 = 6

input_1 = layers.Input(shape=(1, input_dim_1))
input_2 = layers.Input(shape=(input_dim_2,))

bidirectional_gru = layers.Bidirectional(layers.GRU(units=64, return_sequences=False))(input_1)

merged = layers.Concatenate()([bidirectional_gru, input_2])

merged = layers.Dense(64, activation="relu")(merged)
output = layers.Dense(2, activation="softmax")(merged)


model = keras.Model(inputs=[input_1, input_2], outputs=output)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

train_titles = train_titles.reshape(7374, 1, input_dim_1)
train_features = train_features.reshape(7374, input_dim_2)

early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)
history = model.fit([train_titles, train_features],
                    train_labels,
                    epochs=100,
                    batch_size=16,
                    validation_split=0.2,
                    callbacks=[early_stopping])

test_titles = test_titles.reshape(1844, 1, input_dim_1)
test_features = test_features.reshape(1844, input_dim_2)
y_pred = model.predict([test_titles, test_features])
y_pred_classes = y_pred.argmax(axis=1)

print(classification_report(y_true=test_labels, y_pred=y_pred_classes))
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report

input_dim_1 = 12
input_dim_2 = 6

input_1 = layers.Input(shape=(1, input_dim_1))
input_2 = layers.Input(shape=(input_dim_2,))

lstm = layers.LSTM(20, return_sequences=False)(input_1)

lstm = layers.Dropout(0.2)(lstm)


merged = layers.Concatenate()([lstm, input_2])

merged = layers.Dense(64, activation="relu")(merged)

merged = layers.Dropout(0.2)(merged)
output = layers.Dense(2, activation="softmax")(merged)

model = keras.Model(inputs=[input_1, input_2], outputs=output)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

train_titles = train_titles.reshape(7374, 1, input_dim_1)
train_features = train_features.reshape(7374, input_dim_2)

early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)

history = model.fit([train_titles, train_features],
                    train_labels,
                    epochs=100,
                    batch_size=16,
                    validation_split=0.2,
                    callbacks=[early_stopping])

test_titles = test_titles.reshape(1844, 1, input_dim_1)

y_pred = model.predict([test_titles, test_features])
y_pred_classes = y_pred.argmax(axis=1)

print(classification_report(y_true=test_labels, y_pred=y_pred_classes))
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report

input_dim_1 = 12
input_dim_2 = 6

input_1 = layers.Input(shape=(input_dim_1, 1))
input_2 = layers.Input(shape=(input_dim_2,))

conv_layer = layers.Conv1D(filters=64, kernel_size=3, activation='relu')(input_1)
pool_layer = layers.MaxPooling1D(pool_size=2)(conv_layer)
flatten_layer = layers.Flatten()(pool_layer)

merged = layers.Concatenate()([flatten_layer, input_2])

merged = layers.Dense(64, activation="relu")(merged)
output = layers.Dense(2, activation="softmax")(merged)

model = keras.Model(inputs=[input_1, input_2], outputs=output)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

train_titles = train_titles.reshape(7374, input_dim_1, 1)
train_features = train_features.reshape(7374, input_dim_2)

early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)
history = model.fit([train_titles, train_features],
                    train_labels,
                    epochs=100,
                    batch_size=64,
                    validation_split=0.2,
                    callbacks=[early_stopping])

test_titles = test_titles.reshape(1844, input_dim_1, 1)
test_features = test_features.reshape(1844, input_dim_2)
y_pred = model.predict([test_titles, test_features])
y_pred_classes = y_pred.argmax(axis=1)

print(classification_report(y_true=test_labels, y_pred=y_pred_classes))
```
